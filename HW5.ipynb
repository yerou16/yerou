```python
import pandas as pd

df_1 = pd.read_csv('1119.csv') 
df_2 = pd.read_csv('11271.csv') 
df_3 = pd.read_csv('11272.csv') 

print(df_1['text'])
print(df_2['text'])
print(df_3['text'])

text_1 = ' '.join(df_1['text'])
text_2 = ' '.join(df_2['text'])
text_3 = ' '.join(df_3['text'])
```

```python
import re
import string

# 指定需要删除的符号，以逗号和句号为例
symbols_to_remove = ['，', '《','(',')','！','、','「','」','：','？','。','》','.','.%','％','（','）','—','；']

words_to_remove = ["的","在","與","到","和","可能","是","中","月","於","又","有","不","也","就","為","此","其","了","開始","都","已","從","而","著","但","可以","由","年","例如","對","日","上","他","一","說","以","後","來"]

# 创建一个函数来删除符号
def remove_symbols(text):
    for symbol in symbols_to_remove:
        text = text.replace(symbol, '')
    return text

def remove_words(text):
    for word in words_to_remove:
        text = text.replace(word, '')
    return text

def remove_numbers(text):
    text_without_numbers = re.sub(r'\d+', '', text)
    return text_without_numbers


# 在 'text' 列上应用函数来删除符号
df_1['text'] = df_1['text'].apply(remove_symbols)
df_1['text'] = df_1['text'].apply(remove_words)
df_1['text'] = df_1['text'].apply(remove_numbers)

df_2['text'] = df_2['text'].apply(remove_symbols)
df_2['text'] = df_2['text'].apply(remove_words)
df_2['text'] = df_2['text'].apply(remove_numbers)

df_3['text'] = df_3['text'].apply(remove_symbols)
df_3['text'] = df_3['text'].apply(remove_words)
df_3['text'] = df_3['text'].apply(remove_numbers)
# 打印处理后的数据框
print(df_1)
print(df_2)
print(df_3)
```

```python
import jieba

# 加載自定義詞典
with open("custom_dict.txt", "w", encoding="utf-8") as f:
    f.write("蔡英文\n賴清德\n柯文哲\n郭台銘\n九二共識\n聯準會\n侯友宜\n阿爾特曼\n洪都拉斯")

# 你的文本數據
text_1 = ' '.join(df_1['text'])
text_2 = ' '.join(df_2['text'])
text_3 = ' '.join(df_3['text'])

# 中文分詞，添加 HMM 參數
seg_list_1 = jieba.lcut(text_1, cut_all=False, HMM=True)
seg_list_2 = jieba.lcut(text_2, cut_all=False, HMM=True)
seg_list_3 = jieba.lcut(text_3, cut_all=False, HMM=True)

# 將分詞結果轉換為字符串
segmented_text_1 = ' '.join(seg_list_1)
segmented_text_2 = ' '.join(seg_list_2)
segmented_text_3 = ' '.join(seg_list_3)

# 打印分詞後的中文文本
print(segmented_text_1)
print(segmented_text_2)
print(segmented_text_3)
```

```python
from pprint import pprint
from gensim import corpora, models

# 假設 segmented_text_1, segmented_text_2, segmented_text_3 是你的分詞後文本
segmented_texts = [segmented_text_1, segmented_text_2, segmented_text_3]

# 將每個文檔的分詞結果轉換為一個詞彙列表
tokenized_texts = [text.split() for text in segmented_texts]

# 建立字典和文檔-詞頻矩陣
dictionary = corpora.Dictionary(tokenized_texts)
corpus = [dictionary.doc2bow(text) for text in tokenized_texts]

# 建立 LDA 模型
lda_model = models.LdaModel(corpus, num_topics=2, id2word=dictionary, passes=15)

# 打印主題詞
pprint(lda_model.print_topics())
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# 將gensim LDA模型轉換為pyLDAvis格式，並設置sort_topics參數
vis_data = gensimvis.prepare(lda_model, corpus, dictionary, n_jobs=1)

# 將視覺化數據保存到html文件中
pyLDAvis.save_html(vis_data, 'lda_visualization.html')

```
