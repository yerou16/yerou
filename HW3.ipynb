```python
import requests
from bs4 import BeautifulSoup

# 設定要爬取的網址
url = 'https://www.ptt.cc/bbs/KoreaStar/index.html'

# 發送HTTP請求並取得網頁內容
response = requests.get(url)

# 檢查是否成功取得網頁內容
if response.status_code == 200:
    # 解析HTML內容
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 找到標題和連結
    titles = soup.find_all('a')  # 假設連結是用<a>標籤表示的
    for title in titles:
        print(title.text)  # 標題文字
        link = title.get('href')  # 使用get方法來獲取'href'屬性
        print(link) 
else:
    print('無法取得網頁內容，HTTP錯誤碼：', response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# 要搜索的關鍵字
search_keyword = '李俊昊'

url = 'https://www.ptt.cc/bbs/KoreaStar/index.html'

# 處理頁面爬取的函數
def scrape_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        titles = soup.find_all('div', class_='title')
        
        for title in titles:
            if title.a is not None:  # 檢查是否有<a>元素
                title_text = title.a.text
                link = title.a['href']
                if search_keyword in title_text:
                    print(title_text)
                    print('https://www.ptt.cc' + link)
                    print('\n')

    return soup

# 設定要爬取的頁面數
num_pages_to_scrape = float('inf')  # 設為無限大以爬取所有頁面

while num_pages_to_scrape > 0:
    soup = scrape_page(url)
    # 找到上一頁的連結
    prev_page_link = soup.find('a', string='‹ 上頁')
    if prev_page_link:
        url = 'https://www.ptt.cc' + prev_page_link['href']
    else:
        break
    num_pages_to_scrape -= 1

for _ in range(num_pages_to_scrape):
    soup = scrape_page(url)
    # 找到上一頁的連結
    prev_page_link = soup.find('a', string='‹ 上頁')
    if prev_page_link:
        url = 'https://www.ptt.cc' + prev_page_link.get('href')
    else:
        break
```

```python
import requests
from bs4 import BeautifulSoup
import json

# 要搜索的關鍵字
search_keyword = '李俊昊'

url = 'https://www.ptt.cc/bbs/KoreaStar/index.html'

# 創建一個空列表，用於存儲已經印出的留言內容
printed_comments = []

# 處理頁面爬取的函數
def scrape_page(url):
    soup = None  # 初始值設為 None
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        titles = soup.find_all('div', class_='title')
        
        for title in titles:
            if title.a is not None:
                title_text = title.a.text
                link = title.a['href']
                if search_keyword in title_text:
                    article_url = 'https://www.ptt.cc' + link
                    scrape_comments(article_url)

    return soup

# 處理留言爬取的函數
def scrape_comments(article_url):
    response = requests.get(article_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        comments = soup.find_all('span', class_='push-content')
        for comment in comments:
            comment_text = comment.text
            # 檢查留言是否已經印出過，如果沒有，才印出
            comment_identifier = comment_text  # 将 comment_text 作为唯一标识
            if comment_identifier not in printed_comments:
                print(comment_text)
                printed_comments.append(comment_identifier)  # 將留言文本添加到已印出列表中
        print('\n')

# 設定要爬取的頁面數
num_pages_to_scrape = float('inf')

while num_pages_to_scrape > 0:
    soup = scrape_page(url)
    # 找到上一頁的連結
    prev_page_link = soup.find('a', string='‹ 上頁')
    if prev_page_link:
        url = 'https://www.ptt.cc' + prev_page_link.get('href')
    else:
        break
    num_pages_to_scrape -= 1

# 定义要保存为 JSON 文件的文件名
json_file = "comments.json"

# 将 printed_comments 列表保存为 JSON 文件
with open(json_file, "w") as file:
    json.dump(printed_comments, file)

print(f"Data has been saved to {json_file}")
```
